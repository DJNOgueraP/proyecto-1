```python
import pandas as pd
```

```python
#arreglo de filas y columnas
data={"ejex":l1,"ejey":l2}
data
```

```python
#dataframe visualización como excel
df=pd.DataFrame(data)
df
```

```python
#generar una nueva columna
df["calculo"]=df["ejex"]*df["ejey"]
df
```

```python
#visualización de la información-gráficos
import altair as alt
```

```python
#este gráfico no es una captura, puedo mover, hacer zoom para analisis profundos
#presentación dinámica
#plot.lit permite graficar en función del tiempo
grafico1=alt.Chart(df).mark_point(color="red").encode(
    x="ejex",
    y="ejey"
).interactive()
grafico1
```

```python
grafico2=alt.Chart(df).mark_bar(color="red").encode(
    x="ejex",
    y="ejey"
).interactive()
grafico2
```

```python
import matplotlib.pyplot as plt
```
